```python
import random

class Account:
    account_count = 0

    def __init__(self, owner, balance):
        self.bank = "SCì€í–‰"
        self.owner = owner
        self.balance = balance
        self.account_number = self._generate_account_number()

        self.deposit_count = 0
        self.deposit_log = []   # ì…ê¸ˆ ë‚´ì—­ ì €ì¥
        self.withdraw_log = []  # ì¶œê¸ˆ ë‚´ì—­ ì €ì¥

        Account.account_count += 1

    def _generate_account_number(self):
        part1 = f"{random.randint(0, 999):03d}"
        part2 = f"{random.randint(0, 99):02d}"
        part3 = f"{random.randint(0, 999999):06d}"
        return f"{part1}-{part2}-{part3}"

    def deposit(self, amount):
        if amount < 1:
            return False

        self.balance += amount
        self.deposit_log.append(amount)  # âœ… ì…ê¸ˆ ê¸°ë¡
        self.deposit_count += 1

        if self.deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)
            self.balance += interest
            self.deposit_log.append(f"ì´ì(+){interest}")  # ì´ìë„ ê¸°ë¡(ì›í•˜ë©´ ì œê±° ê°€ëŠ¥)

        return True

    def withdraw(self, amount):
        if amount > self.balance:
            return False

        self.balance -= amount
        self.withdraw_log.append(amount)  # âœ… ì¶œê¸ˆ ê¸°ë¡
        return True

    def display_info(self):
        print(
            f"ì€í–‰ì´ë¦„: {self.bank}, ì˜ˆê¸ˆì£¼: {self.owner}, "
            f"ê³„ì¢Œë²ˆí˜¸: {self.account_number}, ì”ê³ : {self.balance:,}ì›"
        )

    # Q10. ì…ê¸ˆ ë‚´ì—­ ì¶œë ¥
    def deposit_history(self):
        print(f"[{self.owner}] ì…ê¸ˆ ë‚´ì—­")
        for item in self.deposit_log:
            print(item)

    # Q10. ì¶œê¸ˆ ë‚´ì—­ ì¶œë ¥
    def withdraw_history(self):
        print(f"[{self.owner}] ì¶œê¸ˆ ë‚´ì—­")
        for item in self.withdraw_log:
            print(item)
```

# Q7. ì´ì ì§€ê¸‰í•˜ê¸° : ì…ê¸ˆ íšŸìˆ˜ê°€ 5íšŒê°€ ë  ë•Œ ì”ê³ ë¥¼ ê¸°ì¤€ìœ¼ë¡œ 1%ì˜ ì´ìê°€ ì”ê³ ì— ì¶”ê°€ë˜ë„ë¡ ì½”ë“œë¥¼ ë³€ê²½í•´ë³´ì„¸ìš”.

```python
import random

class Account:
    account_count = 0

    def __init__(self, owner, balance):
        self.bank = "SCì€í–‰"
        self.owner = owner
        self.balance = balance
        self.account_number = self._generate_account_number()

        self.deposit_count = 0  # ì…ê¸ˆ ì„±ê³µ íšŸìˆ˜(ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜)

        Account.account_count += 1

    def _generate_account_number(self):
        part1 = f"{random.randint(0, 999):03d}"
        part2 = f"{random.randint(0, 99):02d}"
        part3 = f"{random.randint(0, 999999):06d}"
        return f"{part1}-{part2}-{part3}"

    def deposit(self, amount):
        if amount < 1:
            return False

        # 1) ì…ê¸ˆ ì²˜ë¦¬
        self.balance += amount

        # 2) ì…ê¸ˆ íšŸìˆ˜ ì¹´ìš´íŠ¸
        self.deposit_count += 1

        # 3) 5íšŒ ì…ê¸ˆë§ˆë‹¤ ì´ì(1%) ì§€ê¸‰
        if self.deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)  # ì •ìˆ˜ ì› ë‹¨ìœ„ë¡œ ì²˜ë¦¬
            self.balance += interest

        return True

    def withdraw(self, amount):
        if amount > self.balance:
            return False
        self.balance -= amount
        return True

    def display_info(self):
        print(
            f"ì€í–‰ì´ë¦„: {self.bank}, ì˜ˆê¸ˆì£¼: {self.owner}, "
            f"ê³„ì¢Œë²ˆí˜¸: {self.account_number}, ì”ê³ : {self.balance:,}ì›"
        )


# í…ŒìŠ¤íŠ¸
a = Account("íŒŒì´ì¬", 10000)

a.deposit(1000)  # 1
a.deposit(1000)  # 2
a.deposit(1000)  # 3
a.deposit(1000)  # 4
a.deposit(1000)  # 5 -> ì—¬ê¸°ì„œ ì´ì 1% ì§€ê¸‰

a.display_info()
```

```python
import random

# -------------------------
# Q1~Q3ì—ì„œ ë§Œë“  í´ë˜ìŠ¤/í•¨ìˆ˜
# -------------------------
class Character:
    def __init__(self, name: str, level: int, hp: int, atk: int, defense: int):
        self.name = name
        self.level = level
        self.hp = hp
        self.atk = atk
        self.defense = defense

    def is_alive(self) -> bool:
        return self.hp > 0

    def take_damage(self, damage: int) -> int:
        real_damage = damage - self.defense
        if real_damage < 0:
            real_damage = 0
        self.hp -= real_damage
        if self.hp < 0:
            self.hp = 0
        return real_damage

    def attack_target(self, target: "Character") -> int:
        damage = random.randint(1, self.atk)
        return target.take_damage(damage)


class Player(Character):
    def __init__(self, name: str):
        super().__init__(name=name, level=1, hp=100, atk=25, defense=5)
        self.experience = 0

    def gain_experience(self, amount: int):
        self.experience += amount

    def level_up(self):
        if self.experience >= 50:
            self.level += 1
            self.atk += 10
            self.defense += 5


class Monster(Character):
    def __init__(self, name: str, level: int):
        hp = random.randint(10, 30) * level
        atk = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name=name, level=level, hp=hp, atk=atk, defense=defense)


def battle(player: Player, monster: Monster):
    turn = 1
    print(f"\nâš”ï¸ ì „íˆ¬ ì‹œì‘! {player.name}(Lv.{player.level}) vs {monster.name}(Lv.{monster.level})")

    while player.is_alive() and monster.is_alive():
        print(f"\n--- í„´ {turn} ---")

        # 1) í”Œë ˆì´ì–´ ê³µê²©
        raw_damage = random.randint(1, player.atk)  # êµ´ë¦° ë°ë¯¸ì§€(1~ê³µê²©ë ¥)
        dealt = monster.take_damage(raw_damage)     # ì‹¤ì œ ë“¤ì–´ê°„ í”¼í•´(ë°©ì–´ë ¥ ë°˜ì˜)
        print(f"{player.name}ì˜ ê³µê²©! êµ´ë¦¼:{raw_damage} â†’ ì‹¤ì œí”¼í•´:{dealt} | {monster.name} HP: {monster.hp}")

        # ëª¬ìŠ¤í„° ì‚¬ë§ ì²´í¬
        if not monster.is_alive():
            print(f"{monster.name} ì²˜ì¹˜!")
            break

        # 2) ëª¬ìŠ¤í„° ë°˜ê²©
        raw_damage = random.randint(1, monster.atk)
        dealt = player.take_damage(raw_damage)
        print(f"{monster.name}ì˜ ë°˜ê²©! êµ´ë¦¼:{raw_damage} â†’ ì‹¤ì œí”¼í•´:{dealt} | {player.name} HP: {player.hp}")

        turn += 1

    # ì „íˆ¬ ê²°ê³¼ ì²˜ë¦¬
    if player.is_alive():
        exp_gain = monster.level * 20
        player.gain_experience(exp_gain)
        player.level_up()
        print(f"\nâœ… ì „íˆ¬ ìŠ¹ë¦¬! EXP +{exp_gain}")
    else:
        print("\nğŸ’€ ì „íˆ¬ íŒ¨ë°°..")


# -------------------------
# âœ… ëª¬ìŠ¤í„° Në§ˆë¦¬ ìƒì„± + ìˆœì„œëŒ€ë¡œ ìë™ì‚¬ëƒ¥
# -------------------------
def create_monsters(n: int, base_level: int = 1) -> list[Monster]:
    """
    ëª¬ìŠ¤í„° në§ˆë¦¬ ìƒì„±.
    base_levelë¶€í„° ì‹œì‘í•´ì„œ ì ì  ê°•í•´ì§€ê²Œ ë§Œë“¤ì—ˆìŒ(ì›í•˜ë©´ ê³ ì •ë„ ê°€ëŠ¥).
    """
    names = ["ìŠ¬ë¼ì„", "ê³ ë¸”ë¦°", "ëŠ‘ëŒ€", "ìŠ¤ì¼ˆë ˆí†¤", "ì˜¤í¬"]
    monsters = []
    for i in range(n):
        level = base_level + i  # ì ì  ë ˆë²¨ ì¦ê°€ (ê³ ì • ì›í•˜ë©´ base_levelë¡œ ë°”ê¾¸ë©´ ë¨)
        m = Monster(name=f"{random.choice(names)}#{i+1}", level=level)
        monsters.append(m)
    return monsters


def auto_hunt(player: Player, monsters: list[Monster]):
    """
    ëª¬ìŠ¤í„° ë¦¬ìŠ¤íŠ¸ë¥¼ ì•ì—ì„œë¶€í„° ìˆœì„œëŒ€ë¡œ battle ì§„í–‰.
    í”Œë ˆì´ì–´ê°€ ì£½ìœ¼ë©´ ì¦‰ì‹œ ì¤‘ë‹¨.
    """
    for idx, monster in enumerate(monsters, start=1):
        print(f"\n===== [{idx}/{len(monsters)}] {monster.name} (Lv.{monster.level}) ë“±ì¥! =====")
        print(f"í”Œë ˆì´ì–´ ìƒíƒœ: HP {player.hp}, Lv.{player.level}, ATK {player.atk}, DEF {player.defense}, EXP {player.experience}")
        print(f"ëª¬ìŠ¤í„° ìƒíƒœ: HP {monster.hp}, ATK {monster.atk}, DEF {monster.defense}")

        battle(player, monster)

        if not player.is_alive():
            print("\nìë™ì‚¬ëƒ¥ ì¢…ë£Œ: í”Œë ˆì´ì–´ê°€ ì“°ëŸ¬ì¡ŒìŠµë‹ˆë‹¤.")
            return

        print(f"ë‹¤ìŒ ì „íˆ¬ë¡œ ì´ë™! (í˜„ì¬ HP {player.hp}, Lv.{player.level}, EXP {player.experience})")

    print("\nğŸ‰ ëª¨ë“  ëª¬ìŠ¤í„°ë¥¼ ì²˜ì¹˜í–ˆìŠµë‹ˆë‹¤! ìë™ì‚¬ëƒ¥ ì™„ë£Œ!")



```

```python
# ì‹¤í–‰ ì˜ˆì‹œ
# -------------------------
if __name__ == "__main__":
    name = input("í”Œë ˆì´ì–´ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ").strip() or "ìš©ì‚¬"
    player = Player(name)

    N = int(input("ëª¬ìŠ¤í„° ëª‡ ë§ˆë¦¬ ì¡ì„ê¹Œìš”? (N): ").strip() or "5")
    monsters = create_monsters(N, base_level=1)

    auto_hunt(player, monsters)
```
